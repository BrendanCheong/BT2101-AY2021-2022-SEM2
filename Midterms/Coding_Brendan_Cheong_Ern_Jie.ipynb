Please read use the DRUG dataset to answer this question. This is a US county-level dataset. Please carefully read the variable description in Table 4.1. The dataset contains county code–level data on opioid overdose death and the number of electronic prescriptions, along with characteristics of the county population, in the US. The idea is to see whether the number of electronic prescriptions has a negative association with opioid overdose death. Assume that this is a perfectly random sample. Assume that you use opioid overdose death as a dependent variable and the number of electronic prescriptions as the independent variable and use all other variables as control. 

```python
Y = df["Opioid"]
X = df["Electronic_prescriptions"]
model = smf.ols(formula="Opioid ~ Electronic_prescriptions", data=df).fit()
model.summary()
```
